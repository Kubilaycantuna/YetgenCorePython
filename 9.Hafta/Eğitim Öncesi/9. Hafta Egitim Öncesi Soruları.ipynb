# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, brand, model_name, model_year, speed):
        self.brand= brand
        self.model_name= model_name
        self.model_year= model_year
        self.speed= speed  
#
class Bus (Vehicle):
    def __init__(self, brand, model_name, model_year, speed, passenger_capacity= 50):
        super().__init__(brand, model_name, model_year, speed,)
        self.passenger_capacity= passenger_capacity
        print("Bus alt sınıfı oluşturuldu.")
    def intro(self):
        print("""Seçilen Arabanın özellikleri: 
        Marka: {}
        Model İsmi: {}
        Model Yılı: {}
        Hız: {}
        Yolcu Kapasitesi: {} """.format(self.brand, self.model_name,self.model_year,self.speed, self.passenger_capacity))
    def seating_capacity(p=50):
        print(f"The seating capacity of a bus is {p} passengers")
#

Bus1= Bus("Karsan","KRS04", "2020","200",)

Bus1.intro()
Bus1.seating_capacity()

#Bizden istediğiniz çıktıyı alabilmek için seating_capacity adlı fonksiyon oluşturdum fakat 50 çıktısını bir türlü alamadım. Sebebini soracağm.
````
