### Soru 7
YetGen liderlerinden Berkcan, Emirhan ve Eslem, Ramazan Bayramı'nda topladıkları şekerlerini bir araya getirmeyi ve kendi aralarında eşit olarak paylaşmayı kabul ettiler.
Eşit paylaşım yapıldıktan sonra kalan şekerler parçalanacak. Örneğin, eğer toplamda 91 şekerleri varsa , her biri 30 tane alıp kalan 1 tane şekeri parçalayacaklar.
Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
# 7.Sorunun çözümünü buraya yazabilirsin.
berkcanın_şekerleri=float(input("Berkcanın şeker sayısı:"))
emirhanın_şekerleri=float(input("Emirhanın şeker sayısı:"))
eslemin_şekerleri=float(input("Eslemin şeker sayısı:"))

#şeker sayılarını toplayıp mod 3 hesaplanmalı.

print((berkcanın_şekerleri + emirhanın_şekerleri + eslemin_şekerleri)%3)
 #Aslında f" yardımıyla önüne "parçalanacak şeker sayısı" yazmak istedim ama kurgulayamadım. Yardımcı olursanız çok sevinirim.

# sayılar sırasıyla: 34,33,33 şeklinde girildi. 

````
